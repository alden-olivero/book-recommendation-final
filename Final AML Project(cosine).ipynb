```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books=pd.read_csv('BX-Books.csv', sep=";",encoding='latin-1',on_bad_lines="skip" )
books.head()
```

```python
books.columns
```

```python
books=books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-M']]
books.head()
```

```python
books.rename(columns={'Book-Title':'title',"Book-Author":'author',"Year-Of-Publication":"year","Publisher":'publisher',"Image-URL-M":"img_url"},inplace=True)
```

```python
users=pd.read_csv('BX-Users.csv', sep=";",encoding='latin-1',on_bad_lines="skip")
users.head()
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=";",encoding='latin-1',on_bad_lines="skip")
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id',"Book-Rating":'rating'},inplace=True)
ratings.head()
```

```python
print(books.shape)
print(users.shape)
print(ratings.shape)
```

```python
ratings['user_id'].value_counts()
```

```python

x=ratings['user_id'].value_counts()>200
```

```python
y=x[x].index
```

```python
#This expression checks whether each element in the 'user_id' column of the 'ratings' DataFrame is present in the collection 'y'. It returns a boolean Series with the same length as the 'ratings' DataFrame, where each element is either True or False, indicating whether the corresponding user ID in the DataFrame is present in 'y' or not.
ratings=ratings[ratings['user_id'].isin(y)]
ratings.head()
```

```python
rating_with_books=ratings.merge(books,on='ISBN')
rating_with_books.head()
```

```python
#number of ratings each title has received
num_rating=rating_with_books.groupby('title')['rating'].count().reset_index()
num_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
num_rating.head()
```

```python
rating_with_books['rating'] = pd.to_numeric(rating_with_books['rating'])
avg_rating_df = rating_with_books.groupby('title')['rating'].mean().reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'}, inplace=True)
avg_rating_df
```

```python
#final rating merge with main dataframe
popular_df=rating_with_books.merge(num_rating,on='title')
popular_df.head()
```

```python
popular_df=popular_df.merge(avg_rating_df,on='title')
popular_df.head()
```

```python
#if a title receives more rating then we can give correct predictions
popular_df=popular_df[popular_df['num_of_rating']>=100]
popular_df
```

```python
popular_df.sample(10)
```

```python
#if same user ahs given multiple review of same title/book then dropping it
popular_df = popular_df.drop_duplicates('title').sort_values('avg_rating',ascending=False).head(50)

popular_df
```

```python
popular_df
```

```python
x = rating_with_books.groupby('user_id').count()['rating'] > 200
best_users = x[x].index
best_users
```

```python
filtered_rating = rating_with_books[rating_with_books['user_id'].isin(best_users)]
```

```python
y = filtered_rating.groupby('title').count()['rating']>=50
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['title'].isin(famous_books)]
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
book_pivot
```

final_rating: This is the original DataFrame that contains information about ratings and books.

pivot_table(): This is a Pandas function that allows you to create a pivot table from a DataFrame. A pivot table is a way to summarize and aggregate data by rearranging it in a tabular form. It provides a compact representation of the data, allowing you to see the relationship between two or more columns.

columns='user_id': This specifies that the 'user_id' column from the final_rating DataFrame will be used as the column labels (headers) for the pivot table.

index='title': This specifies that the 'title' column from the final_rating DataFrame will be used as the row labels (index) for the pivot table.

values='rating': This specifies that the 'rating' column from the final_rating DataFrame will be used as the values to fill the pivot table.

After executing the pivot_table() function with the provided arguments, the resulting book_pivot will be a new DataFrame containing the pivot table. Each row in the pivot table represents a book title, each column represents a user ID, and the corresponding values represent the ratings given by each user for the respective book.

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

# using kneighbours

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
book_sparse
```

A CSR matrix is a specific sparse matrix representation that is commonly used to store large, sparse matrices efficiently. Sparse matrices are used when the majority of the elements in the matrix are zero, as it saves memory and speeds up computations for certain operations.

```python
#brute force approch with kneighbours classifier is used
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm="brute")
model.fit(book_sparse)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
print(distance,suggestion)
```

distance: This array contains the distances between the specified data point and its k-nearest neighbors. It will be a 1D NumPy array with shape (1, k).
suggestion: This array contains the indices of the k-nearest neighbors in the original dataset book_pivot. It will be a 2D NumPy array with shape (1, k)

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
books_name=book_pivot.index
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestion)):
        books=book_pivot.index[suggestion[i]]
        for j in books :
            print(j)
```

```python
book_name="Message in a Bottle"
recommend_book(book_name)
```

```python
book_name="Zen and the Art of Motorcycle Maintenance: An Inquiry into Values"
recommend_book(book_name)
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores=cosine_similarity(book_pivot)
similarity_scores
```

```python
similarity_scores.shape
```

```python
def recommend2(book_name):
    # index fetch
    index = np.where(book_pivot.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['title'] == book_pivot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['author'].values))
        item.extend(list(temp_df.drop_duplicates('title')['img_url'].values))
        
        data.append(item)
    
    return data
```

```python
recommend2('Message in a Bottle')
```
